```python
from PIL import Image
from imgtools import *
```

```python
my_path='C:/Users/nk/Pictures/Meme/taj.jpg'
```

```python
img = load_image(my_path)
```

```python
display(img)
```

```python
#Dinh vi khu vuc cat
region = (50,100,200,200)
#Cat vung anh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dn
# Xc nh v tr cn dn
paste_position = (100,100)
img.paste(cropped_image, paste_position)
display(img)
```

```python
# Xoay nh
rotated_image = img.rotate(90)
display(rotated_image)
```

```python

rotated_image = rotated_image.rotate(90)
display(rotated_image)
```
