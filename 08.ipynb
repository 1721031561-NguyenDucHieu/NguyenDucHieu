```python
import matplotlib.pyplot as plt
```

```python
#Ve bieu do: line chart
x=[1,2,3,4,5]
y=[10,16,20,30,25]

x1=[1,2,3,4,5]
y1=[20,15,8,9,20]

plt.plot(x,y,'bo-', label='Du lieu') #Ve duong 1
plt.plot(x1,y1,marker='o', color='r', linestyle='-', label='Du lieu-1' ) #Ve duong 2
plt.legend()
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
#Ve bieu do cot Bar Chart 
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color='g', alpha=0.6) 
plt.xlabel('Danh muc') 
plt.ylabel('Gia tri') 
plt.show() 
```

```python
# Vẽ biểu đồ hình tròn Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') 
plt.title('Bieu do hinh tron')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  
plt.title('Biểu đồ hộp ví dụ') 
plt.ylabel('Giá trị')  
plt.show()  
```
